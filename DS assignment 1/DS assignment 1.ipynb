```python
#Attack high   Attack low
attack_mean=pk.Attack.mean()
def set_attack(val):
    if val<attack_mean:
        return "Attack Low"
    elif val==attack_mean:
        return "Attack Neutral"
    else:
        return 'Attack high'
```

```python
pk["Attack High-low"]=pk["Attack"].apply(set_attack)
```

```python
#HP high   HP low
hp_mean=pk.HP.mean()
def set_HP(val):
    if val<hp_mean:
        return "HP Low"
    elif val==hp_mean:
        return "HP Neutral"
    else:
        return 'HP high'
    
pk["HP High-low"]=pk["HP"].apply(set_HP)

pk
```

```python
#speed high  speed low
speed_mean=pk.Speed.mean()
def set_speed(val):
    if val<speed_mean:
        return "speed Low"
    elif val==speed_mean:
        return "speed Neutral"
    else:
        return 'speed high'
    
pk["Speed High-low"]=pk["Speed"].apply(set_speed)

pk
```
